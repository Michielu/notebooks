### Getting current workzones in good order

```python
import pandas as pd
import numpy as np
import json
import requests


dirpath = 'U:/Documents/notebooks/data/'
filename = 'sensorlist.csv'
```

```python
def current_wz(df):
    d = []
    for i in range(len(df)):
        group = df.iloc[i]['Group']
        geo = [df.iloc[i]['Latitude'], df.iloc[i]['Longitude']]
        isFound = False

        for el in d:
            if (el['group'] == group):
                el['geometry'].append(geo)
                isFound = True
                break
            
        if isFound == False:
            x={
                'group':group,
                'geometry':[geo]
            }
            
            d.append(x)
               
         
    return d
```
